```python
import matplotlib.pyplot as plt
x = range(0,10)
y = range(0,10)
fig, axs = plt.subplots(nrows=1, ncols=2)
print(axs)
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2)
axs[0].plot(x,y)
plt.show()
```

```python
fig, axs = plt.subplots(nrows=2, ncols=1)
axs[0].plot(x,y)
plt.show()
```

```python
y2 = list(reversed(range(10)))
fig, axs = plt.subplots(nrows=1, ncols=2)
axs[0].plot(x,y)
axs[1].plot(x,y2)
plt.show()
```

```python
y2 = list(reversed(range(10)))
fig, axs = plt.subplots(nrows=2, ncols=1)
axs[0].plot(x,y)
axs[1].plot(x,y2)
plt.show()
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2)
axs[0].plot(x,y)
axs[0].set_title('Positive Correlation')

axs[1].plot(x,y2)
axs[1].set_title('Negative Correlation')

fig.suptitle('Two types of perfect correlation')
plt.show()
```
